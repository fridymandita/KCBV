#Tranformasi Geometri


```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('cat.jpg')

dstSkala = cv2.resize(img, None, fx=2.5, fy=2, interpolation=cv2.INTER_CUBIC)
cv2_imshow(img)
cv2_imshow(dstSkala)
```

```python
# translasi

import numpy as np

print(img.shape)

baris, kolom, ghgh = img.shape

MTranslasi = np.float32([
     [1, 0, 50],
     [0, 1, 25]])

print(MTranslasi, '\n')


dst = cv2.warpAffine(img, MTranslasi, (baris, kolom))

cv2_imshow(dst)
```

```python
# rotasi

MRotasi = cv2.getRotationMatrix2D((kolom/2, baris/2), 90, 1)

print(MRotasi)

dstRotasi = cv2.warpAffine(img, MRotasi, (kolom, baris))

cv2_imshow(dstRotasi)
```

```python
# Affine

ttkAsal = np.float32([[172,109], [282,65], [272,316]])
ttkTujuan = np.float32([[53,128], [471,46], [275, 385]])

MAffine = cv2.getAffineTransform(ttkAsal, ttkTujuan)

print(MAffine)

dstAffine = cv2.warpAffine(img, MAffine, (kolom, baris))

cv2_imshow(dstAffine)
```
